## Character Text Splitter

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(
                chunk_size=5,
                chunk_overlap=0,
                separator="")
text = "This is a long text that needs to be split into smaller chunks. Each chunk should be no more than 100 characters long."
chunks = text_splitter.split_text(text)
print(chunks)
```

## Recursive Text Splitter

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter 
text_splitter = RecursiveCharacterTextSplitter(
                chunk_size=20,
                chunk_overlap=0,
                separators=["\n\n", "\n", " ", ""])
text = """This is a long text that needs to be split into smaller chunks. Each chunk should be no more than 100 characters long.
    
    The text splitter should try to split on the largest separator first, and then move down to smaller separators if necessary.
    This way, we can ensure that the chunks are as large as possible while still being within the specified chunk size."""
chunks = text_splitter.split_text(text)
print(chunks)
```
