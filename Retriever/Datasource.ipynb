There are multiple retrievers present in langchain based on the user's use case
<br>
**All the retrievers are Runnable**
<br>
Common Retrieves based on the source of data they are fetching -
1. Wikipedia Retriever : queries the wikipedia API to fetch relevent Wikipedia Articles in the from of `LangChain Documnets`
2. VectorStore Retriever : lets us search and fetch documnets from a vector store based on sen=mantic similarity using vector embeddings.

### Wikipedia Retriever

```python
from langchain_community.retrievers import WikipediaRetriever

retriever = WikipediaRetriever(top_k_results=2)
```

```python
query = "What is the history of the Eiffel Tower?"

docs = retriever.invoke(query)
```

```python
docs[0].page_content 
```

### VectorStore Retriever

```python
from langchain_chroma import Chroma
from langchain_huggingface import HuggingFaceEndpointEmbeddings
from langchain_core.documents import Document
```

```python
docs = [
    Document(page_content="The Eiffel Tower is a wrought-iron lattice tower on the Champ de Mars in Paris, France. It is named after the engineer Gustave Eiffel, whose company designed and built the tower."),
    Document(page_content="IPL is a professional Twenty20 cricket league in India, officially known as the Indian Premier League. It was founded by the Board of Control for Cricket in India (BCCI) in 2007."),
    Document(page_content="The Great Wall of China is a series of fortifications that were built across the historical northern borders of China to protect against invasions and raids. It was constructed over several dynasties, with the most well-known sections built during the Ming Dynasty."),
    Document(page_content="The Amazon Rainforest is a vast tropical rainforest in South America, covering parts of Brazil, Peru, Colombia, and several other countries. It is known for its incredible biodiversity and is home to many unique species of plants and animals."),
]
```

```python
embeddings = HuggingFaceEndpointEmbeddings(
    model = "sentence-transformers/all-MiniLM-L6-v2",
)

vector_store = Chroma.from_documents(
    documents=docs,
    embedding=embeddings,
    collection_name='Retrieves'
)
```

```python
retriever = vector_store.as_retriever(search_kwargs={'k':1})
```

```python
query = "What is IPL and When was it started"
results = retriever.invoke(query)
```
