## Text Loader

```python
from langchain_community.document_loaders import TextLoader

file_path = 'textfile.txt'

loader = TextLoader(
    file_path = file_path,

)

docs = loader.load()

print(docs[0])
```

## PyPDFLoader

```python
from langchain_community.document_loaders import PyPDFLoader

loader_obj = PyPDFLoader(
    file_path='Hands-On_Machine_Learning_with_Scikit-Learn-Keras-and-TensorFlow-2nd-Edition-Aurelien-Geron.pdf'
)
# lazy_load creates a geneator object so we cant call any particular documnet directly
# we need to loop on the generator obj to see each object 
document = loader_obj.lazy_load()

# print('Length of the uploaded document : ', len(document))
c=0
for docs in document:
    c+=1
    if c == 31:
        print(docs)
    
```
