## Custom LLM class

```python
import random
from abc import ABC,abstractmethod
```

```python
class MyRunnableclass(ABC):

    def __init__(self):
        return
    
    @abstractmethod
    def invoke():
        pass
```

```python
class MyAnthropicLLM:

    def __init__(self,model:str):
        print(f"LLM Model Claude {model} initiated")
        return

    def predict(self,prompt:str):
        responses = [
            'Virat Kohli is an Indian Cricketer',
            'New Delhi is the capital of India',
            'The Great Wall of China is visible from space',
            'The Eiffel Tower is located in Paris',
            'Alphabet is the parent company of Google'
        ]
        return {'response': random.choices(responses)}
```

```python
class PromptTemplate:
    
    def __init__(self,template,input_variables,format_instructions=None):
        self.template = template
        self.input_variable = input_variables
        
    def format(self,input_dict):
        return self.template.format(**input_dict)
```

```python
llm = MyAnthropicLLM(model = 'Opus-6.7')
```

```python
prompt = PromptTemplate(
    template = "Tell me in {style} about {topic}",
    input_variables = ['style' , 'topic']
)

formatted_prompt = prompt.format({'style':'short','topic':'Virat Kohli'})
print(formatted_prompt)
```

```python
response = llm.predict(formatted_prompt)
print(response)
```
